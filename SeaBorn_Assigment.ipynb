Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans -- >The five plots that we can use in seaborn library are as follow--

I) Line Plot -- It is the simplest plot ,used for visualization in a graph by showing data or indicating in a line..

II) Count Plot -- A Count Plot Is Used To Show The Counts Of Observations In Each Categorical Bin Using Bars.

III) Bar Plot -- A Bar Chart Is A Way Of Comparing A Set Of Categorical Data.

IV) Pair Plot -- A Pair Plot Pairwise Relationships With Other Columns In The Data Frame And Also Plot Pair Plot With Itself.

V) Scatter Plot -- Scatter Plot Is The Same As A Line Plot, In A Line Plot Instead Of Points Being Joined By Line Segments, The Points Are Shown Individually With A Dot, Circle, Or Any Other Shape.


```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.lineplot(x = iris.sepal_length,y=iris.petal_length) # This an example of Line plot
```

```python
sns.scatterplot(x = iris.petal_width , y = iris.sepal_length) # This is example of scatter plot
```

```python
sns.countplot(x ='species' , data = iris) #Example of count plot
```

```python
sns.barplot(x = [1,2,3,4] , y = [5,6,7,8]) # Example of barplot 
```

```python
sns.pairplot(iris,hue = 'species') # pairPlot
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset("flights")

flights = flights.pivot("month", "year", "passengers")

sns.heatmap(flights, annot=True, fmt="d")
```
